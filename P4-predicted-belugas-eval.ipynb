```python
import pandas as pd
```

```python
predicted_belugas = pd.read_csv('../input/hybrid-model-combined/p2_predicted_individual_beluga_id.csv')
predicted_id_cols = ['predicted_id_1', 'predicted_id_2', 'predicted_id_3', 'predicted_id_4', 'predicted_id_5']
predicted_belugas['combined_pred_ids'] = predicted_belugas[predicted_id_cols].agg(','.join,axis=1)
predicted_belugas['in_top_5'] = predicted_belugas.apply(lambda x: x.individual_id in x.combined_pred_ids, axis=1)
```

```python
predicted_belugas
```

```python
individual_id_accuracy = predicted_belugas['in_top_5'].sum() / len(predicted_belugas['in_top_5'])
print('Accuracy of individual id in predicted top 5: ', individual_id_accuracy)
```
